```python
# The list type is a container that holds a number of other objects, in a given order.
# The list type implements the sequence protocol, and also allows you to add and remove objects from the sequence.
# To create a list, put a number of expressions in square brackets:
#     L = []
#     L = [expression, ...]
#     A = B = []    <=> both names will point to the same list
#     A = []; B = A     <==> both names will point to the same list
#     A = []; B = []      <==> independent lists

# Liste türü, belirli bir sırada bir dizi başka nesneyi tutan bir kaptır.
# Liste türü, sıra protokolünü uygular ve ayrıca sıraya nesne eklemenize ve çıkarmanıza olanak tanır.
# Bir liste oluşturmak için, köşeli parantezler içine bir dizi ifade koyun:
# L = []
# L = [ifade, ...]
# A = B = [] <=> her iki isim de aynı listeye işaret edecektir
# A = []; B = A <==> her iki isim de aynı listeye işaret edecektir
# A = []; B = [] <==> bağımsız listeler
```

```python
# The values stored in a list can be accessed using the slice operator ([] and [ : ])
# with indexes starting at 0 in the beginning of the list and working their way to end -1.
# The plus (+) sign is the list concanetation operator, and the asterisk (*) is the repetition operator

# Bir listede saklanan değerlere dilim operatörü ([] ve [ : ]) kullanılarak listenin başında 0'dan başlayan ve -1 sonuna kadar ilerleyen indekslerle erişilebilir.
# Artı (+) işareti liste birleştirme operatörüdür ve yıldız işareti (*) tekrarlama operatörüdür.
```

```python
list = ["abcd",678, 3.14, "John",162]
tinylist = [123, "John"]

print(list)                # Prints complete list                         # Tüm listeyi yazar
print(list[0])             # Prints the first element of the list         # Listenin birinci elemanını yazar
print(list[1:3])           # Prints element starting from 2nd till 3rd    # Listenin 2'nci elemanından 3'üncü elemanına kadar yazar 
print(list[2:])            # Prints elements starting from 3rd element    # Listenin 3'üncü elemanından başlayarak yazar
print(tinylist*2)          # Prints list two times                        # Listeyi iki kez yazar
print(list + tinylist)     # Prints concatenated list                     # Birleştirilmiş listeyi yazar
```

```python
# Python's "for" and "in" constructs are extremely useful, and the first use of them we'll see is with list.
# The "for" construct --for var in list-- is an easy way to look at each element in a list (or other collection.)
# Do not add or remove from the list during iteration.

# Python'un “for” ve “in” yapıları son derece kullanışlıdır ve bunların ilk kullanımını liste ile göreceğiz.
# “for” yapısı --for var in list-- bir listedeki (veya başka bir koleksiyondaki) her bir öğeye bakmanın kolay bir yoludur.
# Yineleme sırasında listeye ekleme veya listeden çıkarma yapmayın.
```

```python
squares = [1,4,9,16,25]
sum = 0

for num in squares:
    sum += num

print(sum)
```

```python
# LIST TYPE: List Methods

# LİSTE TİPİ: Liste Methodları
```

```python
# Here are some other common list methods.

# * list.append(element)       > Adds a single element to the end of the list. Common Error: Does not return the new list, just modifies the original.
# * list.insert(index,element) > Inserts the element at the given index, shifting elements to the right. 
# * list.extend(list2)         > Adds the elements in list2 to the end of the list. Using + or += on a list is similar to using extend().
# * list.index(element)        > Searches for the given element from the start of the list and returns its index.
# * list.remove(element)       > Searches for the first instance of the given element and removes it (throws ValueError if not present).
# * list.sort()                > Sorts the list in place (does not return it).
# * list.reverse()             > Reverses the list in place(does not return it).
# * list.pop(index)            > Removes and returns the element at the given index. Returns the rightmost element if index is omitted (roughly the opposite of append()).

# İşte diğer bazı yaygın liste yöntemleri.

# * list.append(eleman)        > Listenin sonuna tek bir eleman ekler. Yaygın Hata: Yeni listeyi döndürmez, sadece orijinal listeyi değiştirir.
# * list.insert(index, eleman) > Elemanları sağa kaydırarak, verilen indeksteki elemanı ekler. 
# * list.extend(liste2)        > liste2'deki öğeleri listenin sonuna ekler. Liste üzerinde + veya += kullanmak extend() kullanmaya benzer.
# * list.index(eleman)         > Listenin başından itibaren verilen elemanı arar ve indeksini döndürür.
# * list.remove(eleman)        > Verilen elemanın ilk örneğini arar ve kaldırır (mevcut değilse ValueError atar).
# * list.sort()                > Listeyi yerinde sıralar (geri döndürmez).
# * list.reverse()             > Listeyi yerinde tersine çevirir (geri döndürmez).
# * list.pop(index)            > Verilen indeksteki elemanı çıkarır ve döndürür. İndeks atlanırsa en sağdaki elemanı döndürür (append() işlevinin kabaca tersi).
```

```python
list = ["larry", "curly", "moe"]

# Append "shemp" to the end of the list
list.append("shemp")
print(list)  # Output: ['larry', 'curly', 'moe', 'shemp']

# Insert "xxx" at the beginning of the list (index 0)
list.insert(0, "xxx")
print(list)  # Output: ['xxx', 'larry', 'curly', 'moe', 'shemp']

# Extend the list by adding multiple elements
list.extend(["yyyy", "zzzz"])
print(list)  # Output: ['xxx', 'larry', 'curly', 'moe', 'shemp', 'yyyy', 'zzzz']

# Find the index of the element "curly"
print(list.index("curly"))  # Output: 2 (the index of "curly" in the list)

# Remove the first occurrence of "curly"
list.remove("curly")
print(list)  # Output: ['xxx', 'larry', 'moe', 'shemp', 'yyyy', 'zzzz']

# Pop the element at index 1 (removes "larry")
list.pop(1)
print(list)  # Output: ['xxx', 'moe', 'shemp', 'yyyy', 'zzzz']

# Add additional items (for example, we can append "new_item")
list.append("new_item")
print(list)  # Output: ['xxx', 'moe', 'shemp', 'yyyy', 'zzzz', 'new_item']
```

```python
# DATA TYPE: tuple

# VERİ ÇEŞİDİ: tuple
```

```python
# A tuple is another sequence data type that is similar to the list.
# A tuple consist of a number of values separated by commas.
# Unlike lists, however, tuples are enclosed within parantheses.

# The main differences between lists and tuples are:
# * Lists are enclosed in brackets "[]" and their elements and size can be changed, while tuples are enclosed in parantheses () and cannot be updated.
# * Tuples can be thought of as read-only lists.

# Bir tuple, listeye benzeyen başka bir dizi veri türüdür.
# Bir tuple, virgülle ayrılmış bir dizi değerden oluşur.
# Ancak listelerden farklı olarak, tuple'lar parantez içine alınır.

# Listeler ve tuple'lar arasındaki temel farklar şunlardır:
# Listeler parantez “[]” içine alınır ve öğeleri ve boyutları değiştirilebilirken, tuple'lar parantez () içine alınır ve güncellenemez.
# * Tuple'lar salt okunur listeler olarak düşünülebilir.
```

```python
tuple = ("abcd", 678, 3.14, "John", 162)
tinytuple = (123, "John")

print(tuple)                # Prints complete tuple                         # Tüm tuple'ı yazar
print(tuple[0])             # Prints the first element of the tuple         # Tuple'ın birinci elemanını yazar
print(tuple[1:3])           # Prints element starting from 2nd till 3rd     # Tuple'ın 2'nci elemanından 3'üncü elemanına kadar yazar 
print(tuple[2:])            # Prints elements starting from 3rd element     # Tuple'ın 3'üncü elemanından başlayarak yazar
print(tinytuple*2)          # Prints tuple two times                        # Tuple'ı iki kez yazar
print(tuple + tinytuple)    # Prints concatenated tuple                     # Birleştirilmiş tuple'ı yazar
```

```python
# This example returns the item from index -4 (included) to index -1 (excluded).

# Bu örnek, -4 (dahil) indisinden -1 (hariç) indisine kadar olan öğeyi döndürür.

thistuple = ("apple","banana","cherry","orange","kiwi","melon","mango")
print(thistuple[-4:-1])
```

```python
# TUPLE TYPE: Change Tuple Type

# TUPLE TİPİ: Tuple Tipini Değiştirme
```

```python
# Once a tuple is created, you cannot change it's values.
# Tuples are unchangeable.
# You can convert the tuple into a list, change the list, and convert the list back into a tuple.

# Bir tuple oluşturulduktan sonra değerlerini değiştiremezsiniz.
# Tuple'lar değiştirilemez.
# Tuple'ı bir listeye dönüştürebilir, listeyi değiştirebilir ve listeyi tekrar bir tuple'a dönüştürebilirsiniz.
```

```python
# Convert the tuple into a list to be able to change it.

# Değiştirebilmek için tuple'ı bir listeye dönüştürün.
```

```python
x1 = ("apple", "banana", "cherry")  # Original tuple
my_list = list(x1)  # Convert tuple to list (now using 'my_list' instead of 'list')
my_list[1] = "kiwi"  # Modify the list
x1 = tuple(my_list)  # Convert the modified list back into a tuple

print(x1)  # Output: ('apple', 'kiwi', 'cherry')
```

```python
# DICTIONARY TYPE: Accessing Items

# SÖZLÜK TİPİ: Öğelere Erişim
```

```python
# You can access the items of a dictionary by referring to it's key name, inside square brackets.
# There is also a method called get() that will give you the same result.

# Bir sözlüğün öğelerine erişebilirsiniz köşeli parantez içinde anahtar adına atıfta bulunarak.
# Aynı sonucu verecek get() adlı bir yöntem de vardır.
```

```python
thisdict = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}

x = thisdict["model"]  # Accessing the value for the key 'model' using the [] operator
print(x)  # Print the value of 'model'

x = thisdict.get("model")  # Accessing the value for the key 'model' using the get() method
print(x)  # Print the value of 'model'
```

```python
# DICTIONARY TYPE: Loop

# SÖZLÜK TİPİ: Döngü
```

```python
# You can loop through a dictionary by using a for loop.

# for döngüsü kullanarak sözlükte döngü yapabilirsiniz.
```

```python
# Define the dictionary with key-value pairs
thisdict = {
    "brand" : "Ford",      # Key 'brand' with value 'Ford'
    "model" : "Mustang",   # Key 'model' with value 'Mustang'
    "year" : 1964          # Key 'year' with value 1964
}

# Loop through the dictionary to print only the keys
for x in thisdict:          
    print(x)  # Print each key in the dictionary (brand, model, year)

print("\n")  # Print a blank line for separation

# Loop through the dictionary to print the corresponding values
for x in thisdict:
    print(thisdict[x])  # Print the value associated with each key
```

```python
# Loop through the values of the dictionary and print them
for x in thisdict.values():
    print(x)
```

```python
# Loop through both keys and values, by using the items() function:

# items() fonksiyonunu kullanarak hem anahtarlar hem de değerler arasında döngü oluşturun:
```

```python
# Define the dictionary with key-value pairs
thisdict = {
    "brand" : "Ford",      # Key 'brand' with value 'Ford'
    "model" : "Mustang",   # Key 'model' with value 'Mustang'
    "year" : 1964          # Key 'year' with value 1964
}

for x,y in thisdict.items():
    print(x,y)
```

```python
# Adding an item to the dictionary is done by using a new index key and assigning a value to it.

# Sözlüğe bir öğe eklemek, yeni bir indeks anahtarı kullanarak ve buna bir değer atayarak yapılır.
```

```python
# Define the dictionary with key-value pairs
thisdict = {
    "brand" : "Ford",      # Key 'brand' with value 'Ford'
    "model" : "Mustang",   # Key 'model' with value 'Mustang'
    "year" : 1964          # Key 'year' with value 1964
}

# Add a new key-value pair to the dictionary: key 'color' with value 'red'
thisdict["color"] = "red"

# Print the updated dictionary with the new key-value pair
print(thisdict)
```

```python
# DICTIONARY TYPE: Removing Items

# SÖZLÜK TİPİ: Öğelerin Kaldırılması/Silinmesi
```

```python
# There are several methods to remove items from a dictionary.
# The pop() method removes the item with the specified key name.
# The pop() item method removes the last inserted item. 

# Bir sözlükten öğe kaldırmak için birkaç yöntem vardır.
# pop() yöntemi, belirtilen anahtar adına sahip öğeyi kaldırır.
# pop() item yöntemi, eklenen son öğeyi kaldırır. 
```

```python
# Define the dictionary with key-value pairs
thisdict = {
    "brand" : "Ford",      # Key 'brand' with value 'Ford'
    "model" : "Mustang",   # Key 'model' with value 'Mustang'
    "year" : 1964          # Key 'year' with value 1964
}

# Remove the key 'model' from the dictionary
thisdict.pop("model")    # Removes the key 'model' and its associated value ('Mustang') from the dictionary

# Print the updated dictionary
print(thisdict)  # The 'model' key is removed, so it will print the remaining key-value pairs
```

```python
# Define the dictionary with key-value pairs
thisdict = {
    "brand" : "Ford",      # Key 'brand' with value 'Ford'
    "model" : "Mustang",   # Key 'model' with value 'Mustang'
    "year" : 1964          # Key 'year' with value 1964
}

# Remove the last key-value pair from the dictionary
thisdict.popitem()  # Removes the last key-value pair from the dictionary (in this case, 'year': 1964)

# Print the updated dictionary
print(thisdict)  # The last item ('year': 1964) is removed, so the remaining dictionary is printed
```

```python
# DICTIONARY TYPE: Copy a Dictionary

# SÖZLÜK TİPİ: Sözlük Kopyalama
```

```python
# You cannot copy a dictionary simply by typing dict2 = dict1,
# because: dict2 will only be a reference to dict1, and changes made in dict1 will automatically also be made in dict2.

# Bir sözlüğü sadece dict2 = dict1 yazarak kopyalayamazsınız,
# çünkü: dict2 sadece dict1'e referans olacaktır ve dict1'de yapılan değişiklikler otomatik olarak dict2'de de yapılacaktır.
```

```python
# Define the dictionary with key-value pairs
thisdict = {
    "brand" : "Ford",      # Key 'brand' with value 'Ford'
    "model" : "Mustang",   # Key 'model' with value 'Mustang'
    "year" : 1964          # Key 'year' with value 1964
}

# Create a copy of 'thisdict' and assign it to 'mydict'
mydict = thisdict.copy()  # The 'copy()' method creates a new dictionary with the same key-value pairs as 'thisdict'

# Print the copied dictionary 'mydict'
print(mydict)  # Output: {'brand': 'Ford', 'model': 'Mustang', 'year': 1964}
```

```python
# DICTIONARY TYPE: Nested Dictionary

# SÖZLÜK TİPİ: İç İçe Sözlük
```

```python
# A dictionary can also contain many dictionaries, this is called nested dictionaries.

# Bir sözlük de birçok sözlük içerebilir, buna iç içe sözlükler denir.
```

```python
# Define the dictionary 'myfamily' containing nested dictionaries for each child
myfamily = {
    "child1" : {
        "name" : "Emily",  # The name of the first child
        "year" : 2004      # The birth year of the first child
    },
    "child2" : {
        "name" : "Tobias",  # The name of the second child
        "year" : 2007      # The birth year of the second child
    },
    "child3" : {
        "name" : "Linus",   # The name of the third child
        "year" : 2011      # The birth year of the third child
    }
}

# Print the entire 'myfamily' dictionary
print(myfamily)  # This will print the dictionary with nested dictionaries for each child
```

```python
# Define the dictionary for child1 with name and year
child1 = {
    "name" : "Emily",  # The name of the first child
    "year" : 2004      # The birth year of the first child
}

# Define the dictionary for child2 with name and year
child2 = {
    "name" : "Tobias",  # The name of the second child
    "year" : 2007      # The birth year of the second child
}

# Define the dictionary for child3 with name and year
child3 = {
    "name" : "Linus",   # The name of the third child
    "year" : 2011      # The birth year of the third child
}

# Create the main 'myfamily' dictionary that holds references to the individual child dictionaries
myfamily = {
    "child1" : child1,  # Add child1 dictionary to the family dictionary
    "child2" : child2,  # Add child2 dictionary to the family dictionary
    "child3" : child3   # Add child3 dictionary to the family dictionary
}

# Print the entire 'myfamily' dictionary
print(myfamily)  # This will print the family dictionary with nested child dictionaries
```
