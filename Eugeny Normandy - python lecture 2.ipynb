```python
# Write a python program to find the square root of a complex number.
```

```python
import cmath     # Importing the complex math module.

num=eval(input("Enter a complex number: "))     # num = 1+2j will be entered by user.

num_sqrt = cmath.sqrt(num)     # We use "cmath.sqrt()" function to calculate the complex square root.

print("The square root of {0} is {1:0.4f}+{2:0.4f}j".format(num,num_sqrt.real,num_sqrt.imag))
```

```python
# Write a python program to calculate the area of a triangular which side values of it entered by user.
```

```python
# Where a,b,c are three sides of triangle.

a = eval(input("Enter first side: "))
b = eval(input("Enter second side: "))
c = eval(input("Enter third side: "))

semiperimeter = (a+b+c)/2     # Calculate the semi-perimeter (yarı-çevre uzunluğu).

area = (semiperimeter * (semiperimeter - a) * (semiperimeter-b) * (semiperimeter - c))**0.5     # Calculate the area.

print("The area of the triangle is %0.2f"%area)
```

```python
# Write a python program to compute roots of a quadratic equation when coefficientsa,b and c are known.
```

```python
import math     #Import math module.

a = 1
b = 5
c = 6

d = (b**2) - (4*a*c)     # Calculate the discriminant.

# Find two solutions.

sol1 = (-b - math.sqrt(d)) / (2*a)
sol2 = (-b + math.sqrt(d)) / (2*a)

print("The solution are {0} and {1}".format(sol1,sol2))
```

```python
# Write a python program to convert kilometers to miles and display it.
```

```python
# Taking kilometers input from the user.
kilometers=eval(input("Enter value in kilometers: "))

# Conversion factor.
conv_fac=0.621371

# Calculate miles.
miles = kilometers * conv_fac

print("%0.2f kilometers is equal to %0.2f miles." %(kilometers,miles))
```

```python
# Write a python program to swap (takas) two variables by using a temporary variable.
```

```python
x = 5
y = 10

# Create a temporary variable and swap the values.

temp = x
x = y
y = temp

print("The value of x after swapping: {}".format(x))
print("The value of y after swapping: {}".format(y))
```

```python
# Write a python program to swap (takas) two variables without using a temporary variable.
```

```python
x = 5
y = 10

x,y = y,x

print("x = ",x)
print("y = ",y)
```
