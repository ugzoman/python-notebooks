```python
# Write a function to calculate the area of a circle.

# Bir √ßemberin alanƒ±nƒ± hesaplayan bir fonksiyon yaz.
```

```python
def area_circle(radius):                    # Fonsiyonumun ismi area_circle.
    pi = 3.14                               # Sabit pi deƒüerini tanƒ±mlƒ±yorum.
    area = pi * radius**2                   # √áemberin alan form√ºl√ºn√º tanƒ±mlƒ±yorum.
    return area                             # 'area'ya d√∂nd√ºr.
```

```python
# Write a function to convert from Fahrenheit to Celsius. Then, below the function,
# write code that reads a Fahrenheit value, calls the function to make the conversion, and prints the result.

# Fahrenheit'tan Celsius'a d√∂n√º≈üt√ºrmek i√ßin bir fonksiyon yazƒ±n. Ardƒ±ndan, fonksiyonun altƒ±na, Fahrenheit deƒüerini okuyan, 
# d√∂n√º≈üt√ºrme yapmak i√ßin i≈ülevi √ßaƒüƒ±ran ve sonucu yazdƒ±ran kod yazƒ±n.
```

```python
def convert_Fahrenheit_to_Celsius(f):              # D√∂n√º≈ü√ºm fonksiyonumu tanƒ±mladƒ±m.     
    c = (f-32) * 5 / 9                             # D√∂n√º≈ü√ºme girecek deƒüerin i≈üleme alƒ±nacaƒüƒ± form√ºl√º yazdƒ±m.
    return c                                       # c'yi d√∂nd√ºr. 
```

```python
Fahrenheit = float(input("Enter temperature in Fahrenheit: "))   # Kullanƒ±cƒ± Fahrenheit deƒüerini girsin.
celsius = convert_Fahrenheit_to_Celsius(Fahrenheit)              # convert_Fahrenheit_to_Celsius fonksiyonunun √ßƒ±ktƒ±sƒ± "celsius" olsun.
print("%.2f Fahrenheit is: %.2f Celsius"%(Fahrenheit,celsius))   # input ve output deƒüerlerini ekrana yazdƒ±r.
```

```python
# Write a function to convert from Fahrenheit to Celsius or vise versa based on the user demand.

# Kullanƒ±cƒ± talebine g√∂re Fahrenheit'ten Santigrat'a ya da tam tersi d√∂n√º≈üt√ºrme i≈ülemi yapan bir fonksiyon yaz.
```

```python
import sys  # Importing sys to use sys.exit()

# Define a function to convert temperature
def convert_temperature(conversion_type, t):
    if conversion_type == 1:
        return (t - 32) * 5 / 9
    elif conversion_type == 2:
        return (t * 9 / 5) + 32
    else:
        raise ValueError("Invalid conversion type")
```

```python
# Ask the user for conversion type
type_of_temperature = int(input("Enter 1 for Fahrenheit to Celsius, 2 for Celsius to Fahrenheit: "))

# Check for valid input, exit if it's invalid
if type_of_temperature not in (1, 2):
    print("Wrong input. Please enter 1 or 2.")
    sys.exit()  # Exit the program

# Ask for the temperature value
temperature = float(input("Enter temperature: "))

# Perform conversion
temp = convert_temperature(type_of_temperature, temperature)

# Display result
if type_of_temperature == 1:
    print(f"{temperature:.2f}¬∞F is {temp:.2f}¬∞C")
else:
    print(f"{temperature:.2f}¬∞C is {temp:.2f}¬∞F")
```

```python
# Function to compute the area of a rectangle
def rectangle_area(length, width):
    """
    Returns the area of a rectangle.
    
    Parameters:
    - length (float): The length of the rectangle
    - width (float): The width of the rectangle

    Returns:
    - float: Area of the rectangle
    """
    return length * width
```

```python
# Python also accepts function recursion, which means a defined function can call itself.
# Recursion is a common mathematical and programming concept.
# It means that a function calls itself.
# This has the benefit of meaning that you can loop through data to reach a result.
# The developer should be very careful with recursion as it can be quite easy to slipinto writing a function
# which never terminates,or one that uses excess amounts of memory or processor power.

# Python ayrƒ±ca fonksiyon √∂zyinelemesini de kabul eder, yani tanƒ±mlanmƒ±≈ü bir fonksiyon kendini √ßaƒüƒ±rabilir.
# √ñzyineleme yaygƒ±n bir matematiksel ve programlama kavramƒ±dƒ±r.
# Bir fonksiyonun kendini √ßaƒüƒ±rmasƒ± anlamƒ±na gelir.
# Bunun, bir sonuca ula≈ümak i√ßin veriler arasƒ±nda d√∂ng√º yapabileceƒüiniz anlamƒ±na gelen bir faydasƒ± vardƒ±r.
# Geli≈ütirici √∂zyineleme konusunda √ßok dikkatli olmalƒ±dƒ±r, √ß√ºnk√º asla sonlanmayan ya da a≈üƒ±rƒ± miktarda bellek 
# veya i≈ülemci g√ºc√º kullanan bir fonksiyon yazmak olduk√ßa kolay olabilir.
```

```python
# In this example, tri_recursion() is a function that we have defined to call itself ("recurse").
# We us the k variable as the data, which decrements (-1) every time we recurse.
# The recursion ends when the condition is not greater than 0 (i.e. when it is 0).

# Bu √∂rnekte, tri_recursion() kendi kendini √ßaƒüƒ±rmak (‚Äúrecurse‚Äù) i√ßin tanƒ±mladƒ±ƒüƒ±mƒ±z bir fonksiyondur.
# Veri olarak k deƒüi≈ükenini kullanƒ±yoruz ve bu deƒüi≈üken her √∂zyineleme yaptƒ±ƒüƒ±mƒ±zda (-1) azalƒ±yor.
# ≈ûart 0'dan b√ºy√ºk olmadƒ±ƒüƒ±nda (yani 0 olduƒüunda) √∂zyineleme sona erer.
```

```python
def tri_recursion(k):
    if (k>0):
        result = k + tri_recursion(k-1)   # Recursive call.

    else:
        result = 0                        # # Base case: stop recursion when k == 0

    return result
```

```python
result = tri_recursion(6)
print("Recursion Example Results = {0}".format(result))
```

```python
# üîç What the Code Does:

# This function tri_recursion(k) uses recursion to calculate the sum of all integers from k down to 0.
# So, calling tri_recursion(6) is equivalent to:

# 6+5+4+3+2+1+0 = 21

# If k > 0, the function calls itself with k - 1.
# It keeps doing this until k == 0, at which point it returns 0 and starts unwinding.
# The results are added up as the recursion returns.

# Bu kod ne yapƒ±yor?

# Bu tri_recursion(k) fonksiyonu, k'dan 0'a kadar olan t√ºm tamsayƒ±larƒ±n toplamƒ±nƒ± hesaplamak i√ßin √∂zyineleme kullanƒ±r.
# Yani, tri_recursion(6) √ßaƒürƒ±sƒ± ≈üuna e≈üdeƒüerdir:

# 6+5+4+3+2+1+0 = 21

# Eƒüer k > 0 ise, fonksiyon kendini k - 1 ile √ßaƒüƒ±rƒ±r.
# Bunu k == 0 olana kadar yapmaya devam eder, bu noktada 0 d√∂nd√ºr√ºr ve √ß√∂zmeye ba≈ülar.
# √ñzyineleme d√∂nd√ºk√ße sonu√ßlar toplanƒ±r.
```
