```python
# Loops allow us to repeat a block of code multiple times.
# This is the basis for many sophisticated programming tasks.

# We have two ways to write loops:
# * while loops
# * for loops

# In python, while loops are more general.
# In python because you can always write a for loop using a while loop.

# Döngüler bir kod bloğunu birçok kez tekrarlamamıza izin verir.
# Bu, birçok karmaşık programlama görevinin temelini oluşturur.

# Döngü yazmanın iki yolu vardır:
# * while döngüleri
# * for döngüleri

# Python'da while döngüleri daha geneldir/kapsamlıdır.
# Python'da her zaman bir while döngüsü kullanarak bir for döngüsü yazabilirsiniz.
```

```python
# Steps evaluate any code before while evaluate the while loop condition:
# * if it is True, evaluate the block code, and then repeat the evaluation of the condition.
# * if it is False, end the loop, and continue with the code after the loop.

# Adımlar while'dan önce herhangi bir kodu değerlendirin while döngü koşulunu değerlendirin:
# * True ise, blok kodunu değerlendirin ve ardından koşulun değerlendirmesini tekrarlayın.
# * False ise, döngüyü sonlandırın ve döngüden sonraki kodla devam edin.
```

```python
# General form of a while loop:

#      while condition:
#         block
```

```python
# Print the values 1 to 9 using while loop.

# While döngüsünü kullanarak 1'den 9'a kadar yazdırın.
```

```python
# Write a python code using a while loop to add the non-zero numbers that the user types in.

# Kullanıcının girdiği sıfır olmayan sayıları toplamak için bir while döngüsü kullanarak bir python kodu yazın.
```

```python
end_found = False
total = 0

while not end_found:
    x = int(input("Enter an integer to add (0 to end the program): "))

    if x == 0:
        end_found = True

    else:
        total += x

print(total)
```

```python
# Write a python code using a while loop to continue to the next iteration if i is 5.

# i = 5 ise bir sonraki iterasyona devam etmek için while döngüsü kullanarak bir python kodu yazın.
```

```python
i = 0               # i değeri 0'a eşittir.

while i < 10:       # i 10'dan küçük değer(ler)e sahipken aşağıdaki işlemleri yap.
    i += 1          # i değerinin 1 tamsayı artır.
    if i == 5:      # Eğer i değeri 5'e eşitse:
        continue    # Pas geç / es geç.
    print(i)        # i'yi yazdır.
```

```python
# Python program to find sum and average of N natural numbers using while loop.


# while döngüsü kullanarak N tane doğal sayının toplamını ve ortalamasını bulmak için Python programı yaz.
```

```python
number = int(input("PLEASE ENTER ANY NUMBER: "))               # Kullanıcı sayı girsin.

total = 0                                                      # toplam = 0
i = 1                                                          # i = 1

while i <= number:                                             # i değeri kullanıcının girdiği sayıya küçük-eşit iken:

    total += i                                                 # total = total + i
    i += 1                                                     # i = i + 1

average =  total / number                                      # ortalama = toplam / kullanıcının girdiği sayı

print("THE SUM OF NATURAL NUMBERS FROM 1 TO {0} = {1}".format(number,total))
print("AVERAGE OF NATURAL NUMBERS FROM 1 TO {0} = {1}".format(number,average))
```

```python
# Write a python program to create a grade calculator that puts together the results of 5 tests and averages them
# and gives out the result of the 5 grades.

# 5 test sonucunu bir araya getirip ortalamalarını alan 
# ve 5 notun sonucunu veren bir not hesaplayıcı oluşturmak için bir python programı yazın.
```

```python
total = 0
gradeCount = 0

while gradeCount < 5:
    grade = int(input("What was your score: "))

    if grade < 0 or grade > 100:
        print("IT SHOULD BE A NUMBER FROM 0 TO 100.")

    else:
        gradeCount += 1
        total += grade

        if 93 <= grade <= 100:
            print("A+")

        elif 90 <= grade < 93:
            print("A")

        elif 87 <= grade < 90:
            print("A-")

        elif 84 <= grade < 87:
            print("B+")

        elif 81 <= grade < 84:
            print("B")

        elif 78 <= grade < 81:
            print("B-")

        elif 75 <= grade < 78:
            print("C+")

        elif 72 <= grade < 75:
            print("C")

        elif 69 <= grade < 72:
            print("C-")

        elif 66 <= grade < 69:
            print("D+")

        elif 63 <= grade < 66:
            print("D")

        elif 60 <= grade < 63:
            print("D-")

        elif grade < 60:
            print("F")

average = total / gradeCount

print("AVERAGE: "+str(average))
```

```python
# The tea temperature starts at 100 degrees Celcius. You want it at 85 degrees.
# A chip of ice turns out to lower the temperature one degree each time.
# You test the temperature each time, and also print out the temperature before reducing the temperature.


# Çay sıcaklığı 100 Celcius dereceden başlamaktadır. 85 derecede olmasını istiyorsunuz.
# Bir buz parçasının sıcaklığı her seferinde bir derece düşürdüğü ortaya çıkıyor.
# Her seferinde sıcaklığı test edin ve ayrıca sıcaklığı düşürmeden önce sıcaklığı yazdırın.
```

```python
temperature = 100

while temperature >= 85:
    print(temperature)
    temperature -= 1 

print("The tea is cool enough :)")
```

```python
# The user may want to enter a bunch of lines and not count them all ahead of time.
# This means the number of repetitions would not be known ahead of time.

# Kullanıcı bir grup satır girmek ve hepsini önceden saymamak isteyebilir.
# Bu, tekrar sayısının önceden bilinmeyeceği anlamına gelir.
```

```python
lines = list()
testAnswer = input("Press Y if you want to enter more lines: ")

while testAnswer == "Y":
    line = input("Next line: ")
    lines.append(line)
    testAnswer = input("Press Y if you want to enter more lines: ")

print("Your lines were: ")

for line in lines:
    print(line)
```

```python
# Write a python code to perform a guessing game using a while loop.

# While döngüsü kullanarak bir tahmin oyunu gerçekleştirmek için python kodu yazın.
```

```python
import random

number = random.randint(1,3)

number_of_guesses = 0

while number_of_guesses < 5:
    print("Guess a number between 1 and 3: ")

    guess = input()
    guess = int(guess)

    number_of_guesses += 1

    if guess == number:
        print("You win the game!")
        break

print("{0} was the number".format(number))
```

```python
import random               # random kütüphanesini import et.

number = random.randint(1,3)
number_of_guesses = 0

while number_of_guesses < 5:
    print("Guess a number between 1 and 3: ")

    guess = input()
    guess = int(guess)

    number_of_guesses = number_of_guesses + 1

    if guess == number:
        break

if guess == number:
    print("You guessed the number in "+str(number_of_guesses),"Tries!")

else:
    print("You did not guess number. The number was: "+str(number))
```

```python
# This is a simple example of checking whether a number is prime or not.

# Bu, bir sayının asal olup olmadığını kontrol etmenin basit bir örneğidir.
```

```python
import math

print("Enter a number: ")
number = int(input())

i = 2
prime = True

# if the number is not divisible by any number less than the square root of the number then it is prime.

while i <= int(math.sqrt(number)):
    if number % i == 0:
        prime = False
        break

    i=i+1

if number<2:
    prime = False

if prime:
    print(number,"is a prime number")

else:
    print(number,"is not a prime number")
```

```python
# This is a simple example of checking whether a number is Palindrome or not.

# Bu, bir sayının Palindrom olup olmadığını kontrol eden basit bir örnektir.
```

```python
print("Enter a number: ")
number = int(input())

temp = number
reverse_num = 0

while temp != 0:

    reverse_num = (reverse_num*10) + (temp%10)
    temp = temp // 10

if reverse_num == number:
    print("Yes, It is a Palindrome.")

else:
    print("No, It is not a Palindrome")
```

```python
# This is a simple example of checking whether a number is Armstrong or not.
# 153 is an Armstrong number because 153 = (1 * 1 * 1) + (5 * 5 * 5) + (3 * 3 * 3)

# Bu, bir sayının Armstrong olup olmadığını kontrol etmenin basit bir örneğidir.
# 153 bir Armstrong numarasıdır çünkü 153 = (1 * 1 * 1) + (5 * 5 * 5) + (3 * 3 * 3)
```

```python
import math

print("Enter a number: ")
number = int(input())

# to calculate the number of digits in a number:
number_of_digits = int(math.log10(number)) + 1

sum_arm = 0
temp = number

while temp != 0:
    sum_arm = sum_arm + int(math.pow(temp%10,number_of_digits))
    temp = temp // 10

if sum_arm == number:
    print("Yes, an Armstrong number.")

else:
    print("No")
```
