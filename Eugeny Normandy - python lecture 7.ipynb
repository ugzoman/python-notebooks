```python
# Write a program to prompt the user for hours and rate per hour to compute gross pay.
# Take into account that the factory gives the employee 1.5 times the hourly rate for hours worked above 40 hours.

# Brüt ücreti hesaplamak için kullanıcıdan saat ve saat başına ücret isteyen bir program yazın.
# Fabrikanın 40 saatin üzerinde çalışılan saatler için çalışana saat ücretinin 1,5 katını verdiğini dikkate alın.
```

```python
hours = float(input("Enter hours: "))
rate = float(input("Enter rate: "))

if hours < 40:
    pay = hours*rate

else:
    pay = 40*rate +(hours-40)*1.5*rate

print("Pay {0} dollar".format(pay))
```

```python
# Write a shutting down program:

# * First, define a function, shut_down, that takes one argument "s".
# * Then, if the shut_down function receives an "s" equal to "YES", it should return "Shutting Down".
# * If "s" is equal to "NO", then the function should return "Shutdown Aborted".
# * Finally, if shut_down gets anything other than those inputs, the function should return "Sorry".

# Bir kapatma programı yazın:

# * İlk olarak, bir “s” argümanı alan bir “shut_down” fonksiyonu tanımlayın.
# * Ardından, “shut_down” işlevi “s” değeri “YES” değerine eşitse, “Shutting Down” değerini döndürmelidir.
# * Eğer “s” “NO ‘ya eşitse, fonksiyon ’Shutdown Aborted” döndürmelidir.
# * Son olarak, “shut_down” bu girdilerden başka bir girdi alırsa, işlev “Sorry” (Üzgünüz) döndürmelidir.
```

```python
def shut_down(s):
    if s == "YES":
        return "Shutting Down"

    elif s == "NO":
        return "Shutdown Aborted"

    else:
        return "Sorry"
```

```python
s = "YES"
shut_down(s)
```

```python
s = "NO"
shut_down(s)
```

```python
s = "HOLD ON!"
shut_down(s)
```

```python
# Rewrite your pay computation program with time-and-a-half for overtime and
# create a function called "computepay" which takes two parameters (hours and rate).

# Ücret hesaplama programınızı fazla mesai için zaman ve bir buçuk ile yeniden yazın ve
# iki parametre (saat ve ücret) alan “computepay” adında bir fonksiyon oluşturun.
```

```python
def computepay(hours,rate):
    if hours < 40:
        pay = hours*rate
        return pay

    else:
        pay = 40 * rate + (hours - 40) * 1.5 * rate
        return pay
```

```python
hours = float(input("Enter hours: "))
rate = float(input("Enter rate: "))
salary = computepay(hours,rate)

print("Payment is {0}".format(salary))
```

```python
# Follow the steps:
# 1. Define a function called "cube" that takes an argument called "number".
# 2. Make that function return the cube of that number (i.e. that number multiplied by itself and multiplied by itself once again).
# 3. Define a secon function called "by_three" that takes an argument called "number". If that number is divisible by 3,
# "by_three" should call cube(number) and return it's result. 
# 4. Otherwise "by_three" should return "False".

# Adımları takip edin:
# 1. “cube” adında, ‘number’ adında bir argüman alan bir fonksiyon tanımlayın.
# 2. Bu fonksiyonun o sayının küpünü döndürmesini sağlayın (yani o sayının kendisiyle çarpılması ve bir kez daha kendisiyle çarpılması).
# 3. “by_three” adında, ‘sayı’ adında bir argüman alan ikinci bir fonksiyon tanımlayın. Eğer bu sayı 3'e bölünebiliyorsa,
# “by_three” cube(number)'ı çağırmalı ve sonucunu döndürmelidir. 
# 4. Aksi takdirde “by_three” “False” döndürmelidir.
```

```python
def cube(number):
    return number*number*number

def by_three(number):
    if number % 3 == 0:
        return cube(number)

    else:
        return False
```

```python
# Write a python program to construct the following pattern, using a nested for loop.

# Yuvalanmış bir döngü kullanarak aşağıdaki deseni oluşturmak için bir python programı yazın.

#     *
#     * *
#     * * *
#     * * * *
#     * * * * *
#     * * * *
#     * * *
#     * *
#     *
```

```python
n = 5

for i in range(n):
    for j in range(i):
        print("*", end = "")

    print("")

for i in range(n,0,-1):
    for j in range(i):
        print("*", end = "")
    print("")
```

```python
# First part (increasing number of stars)
for i in range(1, 6):  # From 1 to 5
    for j in range(i):  # Print i stars in each row
        print('*', end=' ')
    print()  # Move to the next line after each row

# Second part (decreasing number of stars)
for i in range(4, 0, -1):  # From 4 to 1
    for j in range(i):  # Print i stars in each row
        print('*', end=' ')
    print()  # Move to the next line after each row
```

```python
# Write a python program that accepts a word from the user and reverse it.

# Kullanıcıdan bir kelimeyi alan ve ters çeviren bir python programı yazın. 
```

```python
word = input("Input a word to reverse: ")

for char in range(len(word) -1, -1, -1):
    print(word[char], end = " ")

print("\n")
```

```python
# Write a python function that takes a list and returns a new list with unique elements of the first list.

#     list: [1,2,3,3,3,3,4,5]
#     Unique list: [1,2,3,4,5]

# Bir listeyi alan ve ilk listenin benzersiz öğelerini içeren yeni bir liste döndüren bir python fonksiyonu yazın.
```

```python
def unique_list(l):
    x = []                     # Create an empty list to store unique elements
    for a in l:                # Iterate through each element in the input list 'l'
        if a not in x:         # Check if the element 'a' is already in the result list 'x'
            x.append(a)        # If it's not, add it to 'x'
    
    return x                   # After the loop, return the list 'x' containing only unique elements
```

```python
print(unique_list([1,2,3,3,3,4,5]))
```
