```python
# A program (computer program) is a series of instructions that can be executed by a computer
# to perform a specific task.

# Bilgisayar programı, belirli bir görevi gerçekleştirmek için
# bilgisayar tarafından çalıştırılabilen bir talimatlar zinciridir.
```

```python
# Computer programming is the process of designing and building an executable computer program
# to accomplish a specific computing results.

# Bilgisayar programcılığı, özel bir hesaplama/bilgi işlem sonucunu gerçekleştirmek için
# çalıştırılabilir bir bilgisayar programının tasarlanma ve inşa edilme sürecidir.
```

```python
# In programming languages, we write these instructions by writing what are called statements.

# Programlama dillerinde talimatları komutlar yaza#rak oluştururuz.
```

```python
# Statements in programming languages are kind of like sentences in English.

# Programlamada komutlar İngilizce'deki  cümleler gibidir.
```

```python
# They use words, numbers and punctuations to express one thought, one individual piece.

# Tek bir düşünceyi, tek bir parçayı ifade etmek için kelimeleri, sayıları ve noktalama işaretlerini kullanırlar.
```

```python
# Most programming statements are pretty short, just a few words.

# Programlama komutlarının çoğu oldukça kısa, birkaç kelime kadardır.
```

```python
# A computer language is a set of instructions (rules) that provides a way of telling a computer what operation to perform.

# Bir bilgisayar dili, bir bilgisayara hangi işlemi gerçekleştireceğini söylemenin bir yolunu sağlayan bir dizi talimattır (kural).
```

```python
# A programming language is a notational system for describing computation in both machine-readable 
# and human-readable form.

# Bir programlama dili, hesaplamayı hem makine tarafından okunabilir hem de insan tarafından okunabilir 
# biçimde tanımlamak için kullanılan bir notasyon sistemidir.
```

```python
# A programming language also words, symbols and rules of grammar.

# Bir programlama dili ayrıca kelimeler, semboller ve gramer kurallarıdır.
```

```python
# The grammatical rules are called as syntax.

# Gramer kuralları sintaks olarak adlandırılır.     Sintaks: sözdizim kuralları, sözdizim
```

```python
# Each programming language has a different set of syntax rules.

# Her programlama dili farklı sintaks kurallarına (yazım kurallarına) sahiptir.
```

```python
# Some languages want each of your statements to end with a semicolon, like ending a sentence in English
# with a period and others don't. You just go to the next line and start writing the next statement.

# Bazı diller ifadelerinizin her birinin noktalı virgülle bitmesini ister, tıpkı İngilizce'de bir cümlenin 
# noktayla bitmesi gibi, bazıları ise bunu yapmaz. Sadece bir sonraki satıra geçip bir sonraki ifadeyi yazmaya başlarsınız.
```

```python
# Some languages are all uppercase, some languages are all lowercase, some languages are just don't care.

# Bazı dillerin hepsi büyük harflerle yazılır, bazıları hep küçük harflerle yazılır. Bazılarıysa umursamaz/farketmez.
```

```python
# CPU the chip, the central processing unit that is the brain of any computer, desktop, laptop, server,phone,
# game console well it does not understand any of programming languages.

# CPU çip, herhangi bir bilgisayarın, masaüstünün, dizüstü bilgisayarın, sunucunun, 
# telefonun, oyun konsolunun beyni olan merkezi işlem birimi, programlama dillerinden hiçbirini anlamaz.
```

```python
# We might informally say when we are programming that we are writing code the computer understands,
# but we are not.

# Kabaca programlama yaparken bilgisayarın anlayacağı bir kod yazdığımızı söyleyebiliriz, ancak öyle yapmıyoruz.
```

```python
# Not really. You see the only thing that chip understands is machine code or machine language.

# Pek sayılmaz. Çipin anladığı tek şeyin makine kodu ya da makine dili olduğunu görüyorsunuz.
```

```python
# So these are the real instructions that run directly on your computer hardware.

# Yani bunlar doğrudan bilgisayar donanımınızda çalışan gerçek talimatlardır.
```

```python
# So the question is "Why do not we just write machine code?".

# Peki neden sadece makine koduyla yazıvermiyoruz?
```

```python
# Well because it is almost impossible to do. Writing a full program in a machine code (low-level language)
# would be like digging a tunel through a mountaint with only a teaspoon.

# Çünkü bunu yapmak neredeyse imkansızdır. 
# Makine kodunda (düşük seviyeli bir dil) tam bir program yazmak, sadece bir çay kaşığı ile bir dağda tünel kazmak gibi olurdu.
```

```python
# Now as you move away form the CPU into what are called higher-level languages, you worry less about the hardware.

# Artık CPU'dan uzaklaşıp yüksek seviyeli diller olarak adlandırılan dillere geçtiğinizde, donanım hakkında daha az endişelenirsiniz.
```

```python
# This code is often easier to write and to share even across different platforms, but it can be slower when running.
# Because these languages are not necessarily optimized directly down to the CPU level.

# Bu kodun yazılması ve farklı platformlar arasında bile paylaşılması genellikle daha kolaydır, ancak çalıştırılırken daha yavaş olabilir.
# Çünkü bu dillerin doğrudan CPU seviyesine göre optimize edilmesi gerekmez.
```

```python
# Having said that, these days speed differences are minimal and we will be focusing on the high-level languages.

# Bunu söyledikten sonra, bu günlerde hız farkları minimum düzeyde ve biz yüksek seviyeli dillere odaklanacağız.
```

```python
# But whatever we write has to be converted down to machine code before it can run.

# Ancak yazdığımız her şey çalışabilmesi için makine koduna dönüştürülmesi gerekir.
```

```python
# In general the closer a language is to machine language code the more difficult it is to write 
# and the more you have to know about the actual hardware.


# Genel olarak bir dil makine dili koduna ne kadar yakınsa, yazılması o kadar zor olur ve 
# asıl donanım hakkında o kadar fazla bilgi sahibi olmanız gerekir.
```
