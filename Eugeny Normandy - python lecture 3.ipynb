```python
# Like other programming languages, there are some flow control statements in python as well.

# Diğer programlama dilleri gibi, python'da da bazı kontrol akış komutları/ifadeleri vardır.
```

```python
# Control flow refers to the other in which the program should be executed.

# Kontrol akışı, programın yürütülme sırasını ifade eder.
```

```python
# Generally, the control flow of a program runs from top to bottom.

# Genelde, bir programın akışı yukarıdan aşağıya doğru çalışır.
```

```python
# However, the control flow statements break the general top-to-bottom order of execution 
# by including decision-making, looping and more.

# Bununla birlikte kontrol akış komutları karar verme, döngü ve daha fazlasını dahil ederek
# yukarıdan aşağıya yürütme sırasını kırar/bozar.
```

```python
# This enables to program to first execute a specific block of code based on the conditions used.

# Bu, programın öncelikle kullanılan koşullara göre belirli bir kod bloğu yürütmesini sağlar.
```

```python
num = int(input("Enter a number: "))

if (num % 2 == 0):
    if (num>10):
        print("{0} is both even and greater than 10".format(num))

    else:
        print("{0} is even but less than 10".format(num))

else:
    print("This number is not even. So point checking further")
```

```python
# Write a python code to convert a numerical grade to a letter grade.
```

```python
score = float(input("Enter a score between 0 to 100: "))

if score >= 90:
    letter = "A"

else:
    if score >= 80:
        letter = "B"

    else:
        if score >= 70:
            letter = "C"

        else:
            if score >= 60:
                letter = "D"

            else:
                if score >= 50:
                    letter = "F"

print(letter)
```

```python
# This conditional statement in python allow us to check multiple statements rather than
# just one or two like we saw in if else statement.

# Python'daki bu koşullu ifade, if ve if else komutlarında gördüğümüz sadece bir ya da iki tane durumu kontrol
# etmekten ziyade çoklu komutları kontrol etmemizi sağlar.
```

```python
# İf the first if condition is true than same as in the previous if and if else statements,
# the program will execute the body of the if statement.

# Eğer ilk if koşulu doğruysa, önceki if ve if else komutlarındaki gibi program if komutunu yürütür.
```

```python
# Otherwise, the program will go to the elif block which basically checks for another if statement.

# Aksi taktirde, program temelde başka bir if komutunu kontrol eden elif bloğuna gider.
```

```python
# Write a python code to convert a numerical grade to a letter grade.
```

```python
score = float(input("Enter a score between 0 to 100: "))

if score >= 90:
    letter = "A"

elif score >= 80:
    letter = "B"

elif score >= 70:
    letter = "C"

elif score >= 60:
    letter = "D"

else:
    letter = "F"

print(letter)
```
