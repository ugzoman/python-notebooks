```python
# Import the tkinter module for creating GUI applications
import tkinter

# Create the main window for the GUI application
window = tkinter.Tk()

# Rename the title of the window (this will appear at the top of the window)
window.title("GUI")

# Create a Label widget (a text display area) and add it to the window
# The text on the label will be "Hello World!"
# The 'pack()' method is used to display the label in the window
label = tkinter.Label(window, text = "Hello World!").pack()

# Start the main loop to display the window and keep it open
window.mainloop()
```

```python
import tkinter
from tkinter import messagebox

# Create the main window
window = tkinter.Tk()
window.title("Sophisticated GUI")
window.geometry("500x400")  # Set a fixed size for the window

# Set a background color for the window
window.config(bg="#f0f0f0")

# Create a top frame for general information
top_frame = tkinter.Frame(window, bg="#ffcccb")
top_frame.grid(row=0, column=0, columnspan=3, pady=20, padx=20, sticky="ew")

# Create a bottom frame for buttons
bottom_frame = tkinter.Frame(window, bg="#c1d3d8")
bottom_frame.grid(row=1, column=0, columnspan=3, pady=20, padx=20, sticky="ew")

# Label with a title
title_label = tkinter.Label(top_frame, text="Welcome to the Advanced GUI!", font=("Arial", 16, "bold"), bg="#ffcccb")
title_label.pack()

# Label showing some additional information
info_label = tkinter.Label(top_frame, text="Please choose an option from the buttons below.", font=("Arial", 12), bg="#ffcccb")
info_label.pack()

# Function to show a message box when button is clicked
def on_button_click(button_name):
    messagebox.showinfo("Button Clicked", f"You clicked {button_name}!")

# Button 1 with icon and color
btn1 = tkinter.Button(bottom_frame, text="Button 1", bg="#4CAF50", fg="white", width=20, height=2, command=lambda: on_button_click("Button 1"))
btn1.grid(row=0, column=0, padx=10, pady=10)

# Button 2 with icon and color
btn2 = tkinter.Button(bottom_frame, text="Button 2", bg="#2196F3", fg="white", width=20, height=2, command=lambda: on_button_click("Button 2"))
btn2.grid(row=0, column=1, padx=10, pady=10)

# Button 3 with icon and color
btn3 = tkinter.Button(bottom_frame, text="Button 3", bg="#FF5722", fg="white", width=20, height=2, command=lambda: on_button_click("Button 3"))
btn3.grid(row=1, column=0, padx=10, pady=10)

# Button 4 with icon and color
btn4 = tkinter.Button(bottom_frame, text="Button 4", bg="#9C27B0", fg="white", width=20, height=2, command=lambda: on_button_click("Button 4"))
btn4.grid(row=1, column=1, padx=10, pady=10)

# Button 5 with icon and color
btn5 = tkinter.Button(bottom_frame, text="Button 5", bg="#FF9800", fg="white", width=20, height=2, command=lambda: on_button_click("Button 5"))
btn5.grid(row=2, column=0, columnspan=2, padx=10, pady=10)

# Function to exit the program
def exit_program():
    window.quit()

# Exit button at the bottom-right
exit_button = tkinter.Button(window, text="Exit", fg="red", bg="#f0f0f0", font=("Arial", 10, "bold"), command=exit_program)
exit_button.grid(row=2, column=2, padx=20, pady=20, sticky="e")

# Run the Tkinter event loop
window.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk  # Import ttk explicitly for themed widgets
from tkinter import messagebox, colorchooser

# Function to handle button clicks and display a message
def on_button_click(button_name):
    messagebox.showinfo("Button Clicked", f"You clicked {button_name}!")

# Function to change the background color of the window
def change_bg_color():
    color_code = colorchooser.askcolor(title="Choose a color")[1]
    if color_code:
        window.config(bg=color_code)

# Function to dynamically update the label with input
def update_label():
    user_input = entry.get()  # Get the text from the entry box
    label.config(text=f"You entered: {user_input}")

# Create the main window
window = tk.Tk()
window.title("Advanced GUI")
window.geometry("600x500")  # Set window size
window.config(bg="#f0f0f0")

# Create a notebook (tabbed interface)
notebook = ttk.Notebook(window)  # Use ttk.Notebook instead
notebook.pack(pady=20, padx=20, fill="both", expand=True)

# Create the first tab
tab1 = tk.Frame(notebook, bg="#e3f2fd")
notebook.add(tab1, text="Home")

# Create the second tab
tab2 = tk.Frame(notebook, bg="#fce4ec")
notebook.add(tab2, text="Settings")

# Tab 1 - Home content
# Create a label in the tab1
label = tk.Label(tab1, text="Welcome to the Advanced GUI!", font=("Arial", 18, "bold"), bg="#e3f2fd")
label.pack(pady=20)

# Create an entry box to get user input
entry = tk.Entry(tab1, font=("Arial", 14), width=30)
entry.pack(pady=10)

# Create a button to update the label dynamically
update_button = tk.Button(tab1, text="Update Label", bg="#4CAF50", fg="white", font=("Arial", 12), command=update_label)
update_button.pack(pady=10)

# Create buttons with icons (we'll use text for now, as tkinter doesn't natively support images easily)
button1 = tk.Button(tab1, text="Click Me (1)", bg="#4CAF50", fg="white", width=20, height=2, command=lambda: on_button_click("Button 1"))
button1.pack(pady=10)

button2 = tk.Button(tab1, text="Click Me (2)", bg="#2196F3", fg="white", width=20, height=2, command=lambda: on_button_click("Button 2"))
button2.pack(pady=10)

# Tab 2 - Settings content
# Create a label in the tab2
settings_label = tk.Label(tab2, text="Settings", font=("Arial", 18, "bold"), bg="#fce4ec")
settings_label.pack(pady=20)

# Create a button to change the background color of the window
color_button = tk.Button(tab2, text="Change Background Color", bg="#FF9800", fg="white", font=("Arial", 12), command=change_bg_color)
color_button.pack(pady=20)

# Create a scrollbar for additional content (just as an example)
scrollable_frame = tk.Frame(tab2, bg="#fce4ec")
scrollable_canvas = tk.Canvas(scrollable_frame)
scrollable_canvas.pack(side="left", fill="both", expand=True)

# Adding a scrollbar
scrollbar = tk.Scrollbar(scrollable_frame, orient="vertical", command=scrollable_canvas.yview)
scrollbar.pack(side="right", fill="y")

scrollable_canvas.configure(yscrollcommand=scrollbar.set)
scrollable_canvas.bind("<Configure>", lambda e: scrollable_canvas.configure(scrollregion=scrollable_canvas.bbox("all")))

# Add some dummy content in the scrollable canvas
scrollable_content = tk.Frame(scrollable_canvas, bg="#fce4ec")
scrollable_canvas.create_window((0, 0), window=scrollable_content, anchor="nw")

for i in range(1, 21):  # Add 20 labels for example
    label = tk.Label(scrollable_content, text=f"Scrollable Item {i}", font=("Arial", 12), bg="#fce4ec")
    label.grid(row=i, column=0, pady=5, padx=5)

scrollable_frame.pack(fill="both", expand=True, padx=20, pady=10)

# Run the Tkinter event loop
window.mainloop()
```

```python
# Izgara sınıfının nasıl çalıştığını anlamak için kontrol düğmesini kullanalım.

# İki kontrol düğmesi tanımlayacak ve her iki düğme için de bir metin belirleyeceksiniz.
# Kontrol düğmelerinin durumu, değer ve offvalue tarafından belirlenirken,
# kontrol düğmesinin geçerli durumu ayrı bir değişkende saklanacak IntVar().
# offvalue = 1 ve onvalue = 0 olduğunda karşılık gelen kontrol düğmesi işaretlenecektir.

# Şimdi grid sınıfına gelince,
# düğmeyi satır = 0 ise => ilk satıra
# satır = 1 ise => ikinci satıra yerleştirecek bir parametre satırı geçireceksiniz.
```

```python
import tkinter
from tkinter import *

# Create the main window
top = tkinter.Tk()

# Create IntVar variables to store the state of checkbuttons
CheckVar1 = IntVar()
CheckVar2 = IntVar()

# First checkbutton (Python)
# 'onvalue' is 1 when checked, 'offvalue' is 0 when unchecked
tkinter.Checkbutton(top, text="Python", variable=CheckVar1, onvalue=1, offvalue=0).grid(row=0, sticky=W)

# Second checkbutton (C)
# 'onvalue' is 1 when checked, 'offvalue' is 0 when unchecked
tkinter.Checkbutton(top, text="C", variable=CheckVar2, onvalue=1, offvalue=0).grid(row=1, sticky=W)

# Run the Tkinter event loop
top.mainloop()
```

```python
# Binding or Command Functions

# Bağlama veya Komut İşlevleri
```

```python
# In Tkinter, binding and command are two important concepts that are used to connect user interactions 
# (such as clicks, key presses, etc.) with specific functions or actions.

# 1. Command Functions:
# The command option is used with widgets like buttons, checkbuttons, and others to specify a function 
# that is executed when the widget is activated (e.g., when a button is clicked). 
# This is the most common way of associating a function with a widget.

# Tkinter'de bağlama ve komut, kullanıcı etkileşimlerini birbirine bağlamak için kullanılan iki önemli kavramdır 
# (tıklamalar, tuşa basmalar vb.) belirli işlevler veya eylemlerle.

# 1. Komut Fonksiyonları:
# Komut seçeneği düğmeler, onay düğmeleri ve diğerleri gibi widget'larla birlikte bir fonksiyon belirtmek için kullanılır 
# widget etkinleştirildiğinde (örneğin, bir düğmeye tıklandığında) çalıştırılır. 
# Bu, bir fonksiyonu bir pencere öğesiyle ilişkilendirmenin en yaygın yoludur.
```

```python
import tkinter

def on_button_click():
    print("Button clicked!")

# Create the main window
window = tkinter.Tk()

# Create a button and link the on_button_click function to the button
button = tkinter.Button(window, text="Click Me", command=on_button_click)
button.pack()

# Run the Tkinter event loop
window.mainloop()
```

```python
# In this example:

# The command option is set to the function on_button_click.
# When the button is clicked, the on_button_click() function is executed, and "Button clicked!" is printed to the console.

# Bu örnekte:

# Komut seçeneği on_button_click fonksiyonuna ayarlanmıştır.
# Düğmeye tıklandığında, on_button_click() fonksiyonu çalıştırılır ve konsola “Button clicked!” yazdırılır.
```

```python
# 2. Binding Functions:
# The bind() method is used to associate events (such as a key press, mouse click, etc.) with a function. 
# Unlike the command option (which only works with certain widgets like buttons), 
# bind() allows you to bind any event (such as mouse clicks, key presses, etc.) to a function for any widget or even the window itself.

# Common Event Types:
# "Button-1": Left mouse button click.
# "Button-2": Middle mouse button click.
# "Button-3": Right mouse button click.
# "KeyPress": Key press event.
# "KeyRelease": Key release event.

# 2. Fonksiyonları Bağlama:
# bind() yöntemi, olayları (tuşa basma, fare tıklaması vb.) bir fonksiyonla ilişkilendirmek için kullanılır. 
# Komut seçeneğinin aksine (yalnızca düğmeler gibi belirli widget'larla çalışır), bind() herhangi bir olayı 
# (fare tıklaması, tuşa basma vb.) herhangi bir widget ve hatta pencerenin kendisi için bir fonksiyona bağlamanıza olanak tanır.

# Yaygın Olay Türleri:
# “Düğme-1": Sol fare düğmesi tıklaması.
# “Düğme-2": Orta fare düğmesi tıklaması.
# “Düğme-3": Sağ fare düğmesi tıklaması.
# “KeyPress": Tuşa basma olayı.
# “KeyRelease": Tuş bırakma olayı.
```

```python
# Example: Binding a Mouse Click to a Function

# Örnek: Bir Tuşa Basmayı Bir Fonksiyona Bağlama

import tkinter

def on_click(event):
    print(f"Mouse clicked at ({event.x}, {event.y})")

# Create the main window
window = tkinter.Tk()

# Bind a mouse click event to the window
window.bind("<Button-1>", on_click)  # Left mouse click

# Run the Tkinter event loop
window.mainloop()

# In this example:
# window.bind("<Button-1>", on_click): This binds the left mouse button click event (Button-1) to the on_click function.
# The function on_click(event) receives an event object, which contains information about the event (such as the mouse position).
# When you click on the window, it will print the mouse click coordinates.

# Bu örnekte:
# window.bind(“<Button-1>”, on_click): Bu, sol fare düğmesi tıklama olayını (Button-1) on_click fonksiyonuna bağlar.
# on_click(event) fonksiyonu, olay hakkında bilgi içeren (fare konumu gibi) bir olay nesnesi alır.
# Pencereye tıkladığınızda, fare tıklama koordinatlarını yazdıracaktır.
```

```python
# Example: Binding a Key Press to a Function

# Örnek: Bir Tuşa Basmayı Bir Fonksiyona Bağlama

import tkinter

def on_key_press(event):
    print(f"Key pressed: {event.char}")

# Create the main window
window = tkinter.Tk()

# Bind a key press event to the window
window.bind("<KeyPress>", on_key_press)

# Run the Tkinter event loop
window.mainloop()

# In this example:
# window.bind("<KeyPress>", on_key_press): This binds the key press event (KeyPress) to the on_key_press function.
# The event.char gives the character of the key pressed.
# When you press a key while the window is focused, it will print the key you pressed.

# Bu örnekte:
# window.bind(“<KeyPress>”, on_key_press): Bu, tuşa basma olayını (KeyPress) on_key_press fonksiyonuna bağlar.
# event.char basılan tuşun karakterini verir.
# Pencere odaklanmış durumdayken bir tuşa bastığınızda, bastığınız tuşu yazdıracaktır.
```

```python
# Difference Between bind() and command:

# command:
# Used for widgets like buttons, checkbuttons, and menus to specify a function when the widget is activated (clicked).
# Simpler and more straightforward for button-like interactions.
# Automatically triggers the function when the widget is activated.

# bind():
# Provides more flexibility as it can bind any event (mouse, keyboard, etc.) to any widget or the entire window.
# Can be used to handle complex events like mouse movements, keyboard presses, etc.
# Requires handling of event objects passed to the function.


# bind() ve command arasındaki fark:

# command:
# Düğmeler, onay düğmeleri ve menüler gibi widget'lar için widget etkinleştirildiğinde (tıklandığında) bir fonksiyon belirtmek için kullanılır.
# Düğme benzeri etkileşimler için daha basit ve anlaşılırdır.
# Widget etkinleştirildiğinde fonksiyonu otomatik olarak tetikler.

# bind():
# Herhangi bir olayı (fare, klavye, vb.) herhangi bir pencere öğesine veya tüm pencereye bağlayabildiği için daha fazla esneklik sağlar.
# Fare hareketleri, klavyeye basma vb. gibi karmaşık olayları işlemek için kullanılabilir.
# Fonksiyona aktarılan olay nesnelerinin işlenmesini gerektirir.
```

```python
# Bağlama ve Komut fonksiyonları, bir olay meydana geldiğinde ya da tetiklendiğinde çağrılan işlevlerdir.

# Bağlama fonksiyonlarını anlamak için bir örnek yapalım:
# Tıklandığında DataCamp_Tutorial adlı bir fonksiyonu çağıran bir düşme tanımlayın.
# Ayrıca, DataCamp_Tutorial fonksiyonu GUI başlığıyla birlikte sizin tanımladığınız bir etiket yazdırır.

import tkinter  # Import the tkinter library to create the GUI

# Create the main window of the application
window = tkinter.Tk()
window.title("GUI")  # Set the title of the window

# Define the function that will be called when the button is clicked
def DataCamp_Tutorial():
    # Create a label with the text "Python is amazing!" and pack it into the window
    tkinter.Label(window, text="Python is amazing!").pack()

# Create a button labeled "Click Me!" and associate it with the DataCamp_Tutorial function
# When the button is clicked, the DataCamp_Tutorial function will be executed
tkinter.Button(window, text="Click Me!", command=DataCamp_Tutorial).pack()

# Start the Tkinter event loop to make the window appear and respond to user actions
window.mainloop()
```

```python
import tkinter

window = tkinter.Tk()
window.title("GUI")

label = None  # Create a global variable to store the label

def DataCamp_Tutorial():
    global label
    # If the label already exists, destroy it before creating a new one
    if label:
        label.pack_forget()  # This removes the old label from the window
    label = tkinter.Label(window, text="Python is amazing!")
    label.pack()

tkinter.Button(window, text="Click Me!", command=DataCamp_Tutorial).pack()

window.mainloop()
```

```python
# Explanation of the Improvement:
# Global label variable: This variable keeps track of the label so that it can be removed before a new one is added.
# pack_forget(): This method is used to hide the previous label when the button is clicked again. This ensures that only one label is displayed at a time.
# Now, if you click the button multiple times, the label will be replaced with the new one each time instead of stacking multiple labels on top of each other.

# İyileştirme Açıklaması:
# Global etiket değişkeni: Bu değişken, yeni bir etiket eklenmeden önce kaldırılabilmesi için etiketin kaydını tutar.
# pack_forget(): Bu yöntem, düğmeye tekrar tıklandığında önceki etiketi gizlemek için kullanılır. Bu, bir seferde yalnızca bir etiketin görüntülenmesini sağlar.
# Şimdi, düğmeye birden fazla kez tıklarsanız, birden fazla etiketi üst üste yığmak yerine her seferinde etiket yenisiyle değiştirilecektir.
```
