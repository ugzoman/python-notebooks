```python
# FOR LOOPS
```

```python
# The for loop allows you to iterate over all the values in a sequence, performing the same task on each of the elements.

# For döngüsü, her bir öğede aynı görevi yerine getirerek, bir dizideki tüm değerleri yinelemenize olanak sağlar.
```

```python
# In addition, it is possible to iterate over other objects, such as files, objects or dictionaries.

# Ayrıca, dosya nesneleri veya sözlükler gibi diğer nesneler üzerinde yineleme yapmak da mümkün.
```

```python
# The basic form of the for loop is:

# for döngüsünün temel formu aşağıdakine benzerdir:

#     for var in sequence:
#        statement(s)
#     else:
#        statement(s)
```

```python
# As with the if statement, there must be a colon at the very and of the for statement.

# if ifadesinde olduğu gibi, for ifadesinin de sonunda iki nokta üst üste olmalıdır.
```

```python
# The variable 'var' is a "dummy" variable; it is value is defined locally with the for loop, but when the for loop
# is completed, it will be equal to the last element in the sequence being processed.

# 'var' değişkeni bir "kukla" değişkendir. Değeri konumsal olarak for döngüsüyle birlikte tanımlanır,
# ancak for döngüsü tanımlandığında, işlenmekte olan dizideki son öğeye eşit olur.
```

```python
# Unlike the for statement, where the else (or elif) clause is often used,
# the optional else clause of the for loop is rarely used.

# For ifadesinin aksine, else veya elif yantümcesinin sıklıkla kullanıldığı yerlerde
# for döngüsünün isteğe bağlı başka yantümcesi nadiren kullanılır.
```

```python
# The statements following the else statement are executed when the complete sequence
# of iterations defined by the for loop is completed.

# Else ifadesini izleyen ifadeler, for döngüsü tarafından tanımlanan yinelemelerin
# tam sırası tamamlandığında yürütülür.
```

```python
# If a break statement is encountered inside the for loop, however these statements are not carried out,
# and control goes to the next executable statements after the body of the for/else loop.

# For döngüsü içerisinde bir "break" ifadesi ile karşılaşılırsa, bu ifadeler çalıştırılmaz ve kontrol for/else
# döngüsünün gövdesinden sonraki bir çalıştırılabilir komuta geçer.
```

```python
# The for statement in python differs a bit from what you may be used to in C.

# Python'daki for ifadesi, C'de alışık olabileceğinizden farklıdır.
```

```python
# Rather than giving the user the ability to define both the iteration step and halting condition (as C),
# python's for statement iterates over the items of any sequence (a list or a string), 
# in the order that they appear in the sequence.

# Kullanıcıya hem yineleme adımını hem de durma koşulunu (C olarak) tanımlama yeteneği vermek yerine,
# python'un for komutu herhangi bir dizinin öğeleri (liste veya string) dizide göründükleri sırayla yenilenir.
```

```python
# The range() function supplies the numbers from 1 to 100 for the for loop to calculate the sum of these numbers.

# range() fonksiyonu 1'den 100'e kadar olan sayıların toplamını hesaplar.
```

```python
n = 100

sum = 0

for counter in range(1,n+1):
    sum = sum + counter

print("Sum of 1 until %d: %d" %(n,sum))
```

```python
# Let's see how to use range() with for loop to print the square of the even numbers, 
# cube of numbers divided by 3, and power of 5 of numbers divided by 5.

# Çift sayıların karesini, 3 ile bölünen sayıların küpünü ve 5 ile bölünen sayıların 5'in kuvvetini yazdırmak için 
# range() fonksiyonunu for döngüsü ile nasıl kullanacağımızı görelim.
```

```python
for i in range (1,6):
    if (i%2==0):
        square = i**2
        print("square of",i,"is",square)
    
    elif (i%3==0):
        cube = i**3
        print("cube of",i,"is",cube)

    elif (i%5==0):
        fifth = i**5
        print("power of 5 of",i,"is",fifth)
```

```python
# Calculate the average of grades of 5 students.

# 5 öğrencinin notlarının ortalamasını hesapla.
```

```python
sum = 0

for num in range(5):
    grade = int(input("enter a grade: "))
    if (grade >0):
        sum = sum + grade

    else:
        grade = int(input("Enter a positive number please: "))
        sum = sum + grade

average = sum / (num+1)
print(" {0} / {1} is {2}".format(sum, num+1, average))
```

```python
# Calculate the sum and average of multiple user-entered numbers.

# Kullanıcı tarafından girilen birden çok sayının toplamını ve ortalamasını hesaplayın.
```

```python
numbers = input("Enter numbers seperated by space: ")
numberList = numbers.split()
print("All entered numbers",numberList)

# Calculating the sum of all user entered numbers

sum = 0
for num in numberList:
    sum += int(num)
print("Sum of all entered numbers = ",sum)
avg = sum / len(numberList)

print("Average of all entered numbers = ",avg)
```

```python
# Calculation of the pythagorean numbers.
# Three integers satisfying a**2 + b**2 = c**2 are called pythagorean numbers.

# Pisagor Sayılarının hesaplanması.
# a**2 + b**2 = c**2'yi sağlayan üç tamsayı pisagor sayıları olarak adlandırılır.
```

```python
from math import sqrt

n = input("Maximal number? ")

n = int(n)+1

for a in range(1,n):
    for b in range(a,n):
        c_square = a**2 + b**2
        c = int(sqrt(c_square))

        if((c_square - c**2) == 0):
            print(a,b,c)
```

```python
# Calculate the first n Fibonacci numbers.

# Fibonacci sayılarını hesapla.
```

```python
n_fibonacci_numbers = int(input("Enter n: "))
fibonacci_numbers = [1,1]

for n in range(2,n_fibonacci_numbers):
    fibonacci_n = fibonacci_numbers[n-1] + fibonacci_numbers[n-2]
    fibonacci_numbers.append(fibonacci_n)
    print(fibonacci_numbers)
```

```python
# Write the code to compute the sum of a sequence of numbers.

# x0 + x1 + x2 + ... + xn dizisinin toplamını hesaplamak için gereken kodu yaz.
```

```python
N = int(input("Enter maximum number: "))

result = 0

for n in range(1,N+1):
    result = result + n**2

print("Sum of square of {0} = {1}".format(N,result))
```
